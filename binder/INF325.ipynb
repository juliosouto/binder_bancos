<h1>Funções que criam dados aleatórios</h1>

```python
def create_product_details(**kwargs):
    """
    Cria um produto aleatório para um lojista específico.

    :type id_lojista: int
    :param id_lojista: ID do lojista para este produto
    """

    product_category = random.choice(list(types.keys()))
    model = []
    for i in range(5):
        model.append(random.choice(ascii_uppercase))
        if i < 4:
            model.append(random.choice(digits))
    random.shuffle(model)
    
    a = fake.date_between(start_date='-5y', end_date='today')
    b = fake.date_between(start_date='-2y', end_date='-1y')
    
    product = {
        'ativo': True,
        'marca': random.choice(brands),
        'categoria': types[product_category],
        'modelo': ''.join(model),
        'tipo': product_category,
        'idade': random.randint(1, 9), # fake.date_between(start_date='-3y', end_date='-1y'),
        'condicao': random.choice(conditions),
        'valor': random.choice(prices),
        'voltagem': random.choice(voltages),
        'id_lojista': kwargs['id_lojista'],
        'titulo': None,
        'descricao': None,
        'data_criacao': datetime.combine(a, datetime.min.time()),
        'data_atualizacao': datetime.combine(b, datetime.min.time()),
        'polegadas': None,
        'tecnologia': None,
        'sistema_operacional': None,
        'camera': None,
        'conn': None,
        'like': 0,
    }

    if product["tipo"] in dual_voltage_products:
        product["voltagem"] = 'Dual-Voltage'

    titulo = f'{product["tipo"]} {product["marca"]} {product["voltagem"]}'
    descricao = f'{product["tipo"]} {product["marca"]} \
    \nVoltagem: {product["voltagem"]} \nCondicao: {product["condicao"]}'

    product["titulo"] = titulo
    product["descricao"] = descricao

    if product["tipo"] in products_with_screens:
        product["polegadas"] = random.choice(screen_sizes)
        product["tecnologia"] = random.choice(screen_technologies)
        product["titulo"] = f'{product["tipo"]} {product["marca"]} \
            {product["polegadas"]} {product["tecnologia"]}'
        product["descricao"] = f'{product["tipo"]} {product["marca"]} \
            {product["polegadas"]} {product["tecnologia"]} \nVoltagem: {product["voltagem"]} \
                \nCondicao: {product["condicao"]}'
    
    elif product["tipo"] == 'Smartphone':
        product["os"] = random.choice(mobile_os)
        product["camera"] = random.choice(mobile_cameras)
        product["conn"] = random.choice(mobile_conn)
        product["titulo"] = f'{product["tipo"]} {product["marca"]} \
            {product["polegadas"]} {product["tecnologia"]}'
        product["descricao"] = f'{product["tipo"]} {product["marca"]} \
            {product["polegadas"]} {product["tecnologia"]} \nVoltagem: {product["voltagem"]} \
                \nCondicao: {product["condicao"]}'
    
    return product
```

```python
def build_cnpj():
    cnpj = ''.join([ str(random.randint(0, 9)) for i in range(14) ])
    #cnpj = f'{cnpj[0:2]}.{cnpj[2:5]}.{cnpj[5:8]}/000{cnpj[12]}-{cnpj[13:15]}'
    return cnpj


def build_cpf():
    cpf = ''.join([ str(random.randint(0, 9)) for i in range(11) ])
    #cpf = f'{cpf[0:3]}.{cpf[3:6]}.{cpf[6:9]}-{cpf[9:11]}'
    return cpf


def build_phone_number():
    ddds = [ n for n in range(11, 100, 1) ]
    first_digit = [2, 3, 4, 5]
    nums = random.randint(1111111, 9999999)
    nums = str(nums)[:3] + ' ' + str(nums)[3:]
    return f'+55 ({random.choice(ddds)}) {random.choice(first_digit)}{nums}'

companies = {}

for _ in range(companies_qty):

    name = fake.company().split(',')
    name = f'{" ".join(reversed(name))} {random.choice(suffixes)}'
    type_ = random.choice(natures)
    cpf, cnpj = ['', '']

    if type_ == 'JURIDICA':
        cnpj = build_cnpj()
    else:
        cpf = build_cpf()
    
    # Cria os produtos para cada lojista
    products = {}

    for i in range(products_qty):
        products[i] = create_product_details(id_lojista=_+1)

    endereco = fake.address().replace('\n', ' ').split(' ')[:-2]
    a = f'{endereco[1]}' if endereco[1][0] not in digits else ''.strip()
    b = f'{endereco[2]}' if endereco[2][0] not in digits else ''.strip()
    rua = f'{random.choice(logradouros)} {a} {b}'.replace(',', '')
    numero = endereco[0] if endereco[0][0] in digits else endereco[1] is endereco[1][0] in digits
    bairro = endereco[-1].replace(',', '')
    cidade = fake.city()
    estado = random.choice(estados_br)
    cep = random.randint(10000000, 99999999)
    
    complemento = ''
    if 'Suite' in endereco:
        complemento = f'{endereco[endereco.index("Suite")]} {endereco[endereco.index("Suite")+1]}'

    companies[_+1] = {
        'id': _+1,
        'nome': name.strip(),
        'cnpj': cnpj,
        'cpf': cpf,
        'rua': rua,
        'numero': numero,
        'bairro': bairro,
        'cidade': cidade,
        'estado': estado,
        'cep': cep,
        'complemento': complemento,
        'email': fake.company_email(),
        'telefone': build_phone_number(),
        'tipo': type_,
        'data_criacao': fake.date_between(start_date='-5y', end_date='today'),
        'data_atualizacao': fake.date_between(start_date='-2y', end_date='-1y'),
        'produtos': products,
    }
```

<h1>Criando as tabelas no MariaDB</h1>

```python
lojista_schema = """create table lojista
(
    id          bigint                                    auto_increment comment 'Id único da tabela',
    nome        varchar(1000)                             not null comment 'Nome da loja',
    cnpj        varchar(14)                               null comment 'CNPJ do logista',
    cpf         varchar(11)                               null comment 'CPF do logista',
    bairro      varchar(255)                              null comment 'Bairro do lojista',
    cep         varchar(10)                              null comment 'CEP do lojista',
    complemento varchar(255)                              null comment 'Complemento do lojista',
    endereco    varchar(500)                              null comment 'Endereço do lojista',
    numero      varchar(50)                               null comment 'Número do lojista',
    estado      varchar(2)                                null comment 'Estado do lojista',
    email       varchar(255)                              not null comment 'E-mail do logista',
    telefone    varchar(100)                              not null comment 'Telefone do lojista',
    tipo        varchar(255)                              not null comment 'Tipo do cliente Físico/Jurídico',
    data_criacao datetime(3) default CURRENT_TIMESTAMP(3) not null comment 'Data de criação',
    data_atualizacao datetime(3)                          null comment 'Última atualização',
    constraint lojista_pk primary key (id)
);"""


index_schemas = [
    'create unique index lojista_id_uindex on lojista (id);',
    'create index lojista_index_cnpj on lojista (cnpj);',
    'create index lojista_index_cpf on lojista (cpf);',
    'create index lojista_index_data_criacao on lojista (data_criacao);',
    'create unique index produto_id_uindex on produto (id);',
    'create index prod_ix_data_criacao on produto (data_criacao);',
]

    
produto_schema = """create table produto
(
    id               bigint auto_increment comment 'Id único da tabela',
    ativo            boolean default true not null comment 'Produto ativo?',
    idade            int                      null comment 'Idade do produto',
    condicao         varchar(1000)            not null comment 'Condição do produto',
    valor            decimal(10, 2) default 0 not null comment 'Valor do produto',
    id_lojista       bigint                   not null comment 'Id do lojista',
    titulo           varchar(255)            null comment 'Titulo do produto',
    data_criacao     datetime(3) default CURRENT_TIMESTAMP(3) not null comment 'Data de criação',
    data_atualizacao datetime(3)        null comment 'Última atualização',
    constraint produto_pk
        primary key (id),
    constraint produto_lojista_fk
        foreign key (id_lojista) references lojista (id)
            on update cascade on delete cascade
);"""

# Try except somente para quando as tabelas já existem 
try: maria_cursor.execute(lojista_schema)
except: pass
try: maria_cursor.execute(produto_schema)
except: pass

for q in index_schemas:
    try: maria_cursor.execute(q)
    except: pass
```

<h1>Populando as tabelas - Mariadb</h1>

```python
# Nomes das colunas da tabela lojista
lojista_headers = [
    'nome', 'cnpj', 'cpf', 'bairro', 'cep', 'complemento',
    'endereco', 'numero', 'estado', 'email', 'telefone', 'tipo',
    'data_criacao', 'data_atualizacao'
]

# Nomes das colunas da tabela produto
produto_headers = [
    'ativo', 'idade', 'condicao', 'valor', 'id_lojista',
    'titulo', 'data_criacao', 'data_atualizacao'
]


# Tabela lojista
for k, v in companies.items():  # k == id (each row)
    
    lojista_cols_temp = []
    lojista_values_temp = []
    
    for col, value in companies[k].items():
        if col in lojista_headers:
            lojista_cols_temp.append(col)
            lojista_values_temp.append(value)
    
    # Insere cada linha na tabela lojista
    s = f'({"%s," * len(lojista_cols_temp)})'.replace(',)', ')')
    cols = f'({",".join(lojista_cols_temp)})'
    q = f'INSERT INTO lojista {cols} VALUES {s}'
    maria_cursor.execute(q, lojista_values_temp)

# Commit
maria_conn.commit()

    
# Tabela produto
for k, v in companies.items():  # k == id (each row)
    
    for row in companies[k]['produtos'].keys():
        
        produto_cols_temp = []
        produto_values_temp = []
        
        for col, value in companies[k]['produtos'][row].items():
            if col in produto_headers:
                #print(f'{col}: {value}')
                produto_cols_temp.append(col)
                produto_values_temp.append(value)
    
    # Insere cada linha na tabela produto
    s = f'({"%s," * len(produto_cols_temp)})'.replace(',)', ')')
    cols = f'({",".join(produto_cols_temp)})'
    q = f'INSERT INTO produto {cols} VALUES {s}'
    maria_cursor.execute(q, produto_values_temp)

# Commit
maria_conn.commit()
```

<h4>Redis gerenciado pela própria aplicação a partir de dados de produtos no Mongodb</h4>

```python
for k, v in companies.items():  # k == id (each row)
    i = 0
    for product_row, product in companies[k]['produtos'].items():
        redis_client.mset({f"likes_{str(product['_id'])}_{product['categoria']}": random.randint(1, 5)})
        i += 1
        if i == math.ceil(len(companies[k]['produtos'].keys()) / 2):
            break

keys_ = list(redis_client.scan_iter())

d = {}
for k in keys_:
    if 'likes_' in str(k):
        d[k] = int(redis_client.mget(k)[0])
d = dict(sorted(d.items(), key=lambda item: item[1], reverse=True))

categoria = 'Electronics'

best = {}
nota_max = max(d.values())
for k, v in d.items():
    if v >= nota_max:
        _ = k.decode('utf-8').split('_')
        if _[2] == categoria:
            best[_[1]] = _[2]

i = 0
for k, v in best.items():
    print(f'{k}: {v}')
    i += 1
    if i == 10:
        break
#pprint(best)
```
